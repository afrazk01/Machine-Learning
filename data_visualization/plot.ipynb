```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0,5,11)
x
```

```python
y = x**2
```

```python
plt.plot(x,y)
plt.title("test plot")
plt.xlabel("Rand x Values")
plt.ylabel("Rand y value")
```
