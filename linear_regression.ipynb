```python
import pandas as pd
import numpy as np 
import matplotlib.pyplot as plt
from sklearn import linear_model
```

```python
data = pd.read_csv('house_prices.csv')
data.head()
```

```python
%matplotlib inline 
plt.xlabel('area')
plt.ylabel("price")
plt.scatter(data.area,data.prices,color = "red", marker="+")
```

```python
X = data[['area']] 
y = data['prices']  

model = linear_model.LinearRegression()
```

```python

model.fit(X, y)

print("Coefficient (Slope):", model.coef_[0])
print("Intercept:", model.intercept_)
```

```python
y_pred = model.predict(X)
```

```python
plt.scatter(X, y, color='blue', label='Actual Prices')
plt.plot(X, y_pred, color='red', label='Regression Line')
plt.xlabel('Area (sq ft)')
plt.ylabel('Price ($)')
plt.title('Linear Regression: Area vs Price')
plt.legend()
plt.show()
```

```python
model.predict([[3000]])
```

```python
x_ = model.coef_
y_ = model.intercept_

def check_pred(x,m,b):
    return m * x +b

check_pred(3000,x_,y_)
```

```python
data = pd.read_csv("exersise_single_linear_reg.csv")
data.head()
```

```python
%matplotlib inline 
plt.xlabel("Year")
plt.ylabel("Income Per Capital USD")
plt.scatter(data.year,data["per capita income (US$)"], color = 'blue', marker="+")
```
